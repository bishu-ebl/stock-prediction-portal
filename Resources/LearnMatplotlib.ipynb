### Subplots and Legends

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 20, 25, 30, 40]
predicated_price = [8, 21, 25, 28,38]
```

```python
plt.plot(days, original_price, 'b-o', label = 'Original Price'); # b refer color Blue o represent data point
plt.plot(days, predicated_price, 'r-o', label = 'Predicated Price');
plt.legend();
plt.title('Plot with Legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
```

## Subplots

```python
# First subplot - Original Prices
plt.subplot(3, 1, 1) # Rows-3, Cols-1, Index-1
plt.plot(days, original_price, 'b-o', label = 'Original Price');
plt.title('Original prices over time');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend();

# Second subplot - Predicated Prices
plt.subplot(3, 1, 2) # Rows-3, Cols-1, Index-1
plt.plot(days, predicated_price, 'r-o', label = 'Predicated Price');
plt.title('Predicted prices over time');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend();

# Third subplot - The difference between original and predicted prices
price_difference = [o-p for o, p in zip(original_price, predicated_price)]
plt.subplot(3, 1, 3) # Rows-3, Cols-1, Index-1
plt.bar(days, price_difference, color = 'purple', label = 'Price Difference');
plt.title('Difference between original and predicted prices');
plt.xlabel('Days');
plt.ylabel('Price Difference');
plt.legend();

plt.tight_layout()
```

## Saving the figure

```python
plt.plot(days, original_price, 'b-o', label = 'Original Price'); # b refer color Blue o represent data point
plt.plot(days, predicated_price, 'r-o', label = 'Predicated Price');
plt.legend();
plt.title('Plot with Legends');
plt.xlabel('Weekdays');
plt.ylabel('Price'); 
plt.savefig('example_price_predict.png', dpi=300) # to save the image
```

# Step 4 Data Preprocessing

```python
df.shape
```

```python
# Splitting the data into training and testing dataset
data_training = pd.DataFrame(df.Close[0 : int(len(df)*0.7)])
data_testing = pd.DataFrame(df.Close[int(len(df)*0.7) : int(len(df))])
print(data_training)
print(data_testing)
```

```python
data_training
```

```python
data_testing
```

```python
# Scaling down the data between 0 and 1
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
data_training_array = scaler.fit_transform(data_training)
data_training_array
```

```python
type(data_training_array)
```

```python
data_training_array.shape
```

# Step 5 - Data Sequence

```python
# LSTM model require the inputed data in a sequential manner. This is required to predict tomorrows stock price is depended on last few day stock prices
```

```python
x_train = []
y_train = []

for i in range(100, data_training_array.shape[0]): # [0] represent 1st index
    x_train.append(data_training_array[i-100 : i])
    y_train.append(data_training_array[i, 0])

x_train, y_train = np.array(x_train), np.array(y_train)

y_train
```

# Step 6 - Build Machine Learning Model

```python
import tensorflow as tf
print(tf.__version__)
```

```python
# ML Model
from keras.models import Sequential
from keras.layers import Dense, LSTM, Input
```

```python
model = Sequential()
model.add(Input(shape=(100,1))) # the model will add 100 previous days data and give one tomorrows price with 1 feature, where feature nothing but our closign price
model.add(LSTM(units=128, activation='tanh', return_sequences=True)) # This code is used to add layers in the model where units= 128 means initialize LSTM to have 128 neurons activation='tanh' which helps the network to handle both positive and negative value which very require to capture pattern in sequential data, set return_sequence=True beause we want this layer to return full sequence to next LSTM Layer
model.add(LSTM(units=64)) # Add another LSTM model to add sequnce of output from previous LSTM model that means this layer will take sequence of putput from previous LSTM model and generate a single output
model.add(Dense(25)) # Dense layer is typicall used at the end of neural network to retuen the learn feature to the final output. It will take the single output gernerate from previous layer and analyze this output into 25 different angels and then pass the finding into next layer
model.add(Dense(1)) # This dense layer take 25 output from the previous dense layer and combine them to single output. This actually give us the final output price.
```

# Step 7- Model Training

```python
model.compile(optimizer='adam', loss='mean_squared_error' ) # adam is optimizer algoritham that is used to optimize the model
model.fit(x_train, y_train, epochs=50) # epochs= 50 means asking model to go through the x_train ans y_train 50 times, like reading the data 50 times
```

```python
model.summary()
```

```python
# Save the model
model.save('stock_prediction_model.keras')
```

# Step 10 - Model Evalution(to check whether our prediction is correct or not

```python
# Mean Squared Error(MSE)
# MSE measured the average squared difference between the pridicted price and actual values. It shows how inaccurate the model prediction are
# Lower MSC generally means that our model predictional are closer to the actual value.
# Here is the calculation of MSE
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_predicated)
print(f"Mean Squared Error(MSE): {mse}")
```

```python
# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error(RMSE): {rmse}")
```

```python
# R-Squared. It measure how our model prediction match the actual value
r2 = r2_score(y_test, y_predicated) # the value should lies between 0 and 1. Value close 1 means more acurate value
print(f"R-Squared: {r2}")
```
